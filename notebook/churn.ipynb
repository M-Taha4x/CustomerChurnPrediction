__Head Count of Categorical Columns__

```python
for col in num_col:
    print(f"{col}")
    plt.figure(figsize=(10,6))
    sns.countplot(data=df,x=df[col])
    plt.title(f"Number of Observations in {col}")
    plt.show()
```

```python
print(df['City'].value_counts())
```

```python
for col in num_col:
    print(f"{col}")
    plt.figure(figsize=(10,6))
    plt.hist(df[col],bins=30,density=False,cumulative=False,color='steelblue',edgecolor='white')
    plt.title(f"Histogram of {col}")
    plt.show()
```
